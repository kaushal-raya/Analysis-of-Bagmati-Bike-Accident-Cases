<h1 align="center"><b><h1 align="center"><b>Plots Using Matplotlib</b></h1>
 </b></h1>


```python
district_stats = df.groupby('District')['Total Casualties'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
district_stats.plot(kind='bar')
plt.title('Total Casualties by District')
plt.ylabel('Casualties')
plt.tight_layout()
plt.show()
```

```python
weather_stats=df.groupby('Weather') ['Total Casualties'].sum().sort_values(ascending=False)
plt.figure(figsize=(15,5))
weather_stats.plot(kind ='pie')
plt.title('Total Casualties by Weather')
plt.tight_layout()
plt.show()
```

```python
cause_stats=df.groupby('Cause')['Total Casualties'].sum().sort_values(ascending=False)
plt.figure(figsize=(15,9))
cause_stats.plot(kind='bar')
plt.title('Top 5 causes of Bike Accidents  ')
plt.ylabel('Total Casualties')
plt.xlabel('Cause')
plt.tight_layout()
plt.show()
```

<h1 align="center"><b>Conclusion</b></h1>



This project has revealed several critical patterns in bike accident cases within Bagmati Province from 2022 to 2024. The data shows a clear upward trend in accident cases, with 2022 experiencing particularly high rates. Key contributing factors identified include over-speeding and drunk driving, both of which remain significant challenges for road safety. Furthermore, Kathmandu stands out as the district with the highest number of recorded accidents, followed by Kavrepalanchowk, highlighting the need for focused interventions in these areas.

The findings from this analysis provide valuable, actionable insights for policymakers, local authorities, and NGOs. By understanding the main causes and hotspots of bike accidents, stakeholders can design more effective safety campaigns, enforce stricter traffic regulations, and prioritize infrastructure improvements in the most affected regions. Ultimately, these targeted efforts can help reduce accident rates, save lives, and create safer roads for all residents of Bagmati Province.
